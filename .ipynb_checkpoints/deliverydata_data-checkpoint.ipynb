## datetime 형변환

```python
delivery_data['datetime'] = delivery_data['datetime'].apply(pd.to_datetime)
```

## datetime 년월일 분리
- float() argument must be a string or a number, not 'Timestamp'

```python
delivery_data['year'] = delivery_data['datetime'].apply(lambda x : x.year)
delivery_data['month'] = delivery_data['datetime'].apply(lambda x : x.month)
delivery_data['day'] = delivery_data['datetime'].apply(lambda x : x.day)
```

```python
delivery_data = delivery_data.loc[:,['year','month','day','Day of the week','city','population','hour', 'rain','dust','humidity','precipitation','temp','windspeed','holiday','sectors']]
```

## 요일 원핫 인코딩

```python
# 레이블링 인코딩
```

```python
label_delivery_data = delivery_data.copy()
```

```python
label_length = len(label_delivery_data)
```

```python
label_delivery_data['Day of the week'].unique()
```

```python
for i in range(label_length):
    if label_delivery_data['Day of the week'][i] == 'Monday':
        label_delivery_data['Day of the week'][i] = 1
    elif label_delivery_data['Day of the week'][i] == 'Tuesday':
        label_delivery_data['Day of the week'][i] = 2
    elif label_delivery_data['Day of the week'][i] == 'Wednesday':
        label_delivery_data['Day of the week'][i] = 3
    elif label_delivery_data['Day of the week'][i] == 'Thursday':
        label_delivery_data['Day of the week'][i] = 4
    elif label_delivery_data['Day of the week'][i] == 'Friday':
        label_delivery_data['Day of the week'][i] = 5
    elif label_delivery_data['Day of the week'][i] == 'Saturday':
        label_delivery_data['Day of the week'][i] = 6
    elif label_delivery_data['Day of the week'][i] == 'Sunday':
        label_delivery_data['Day of the week'][i] = 7
```

```python
label_delivery_data
```

```python
# 원 핫 인코딩
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder(sparse=False)
ohe.fit(delivery_data[['city']])
city_list = list(delivery_data['city'].unique())
Onehot = pd.DataFrame(ohe.transform(delivery_data[['city']]), columns=city_list, dtype='int64')
```

```python
Onehot
```

## 데이터 분리
- 레이블링
- 원 핫

*dust 값은 결속치로 제거*

```python
columns_list = ['year','month','day','Day of the week', 'city', 'population', 'hour','rain', 'humidity', 'precipitation', 'temp', 'windspeed','holiday']
X_delivery_data = label_delivery_data[columns_list]
```

```python
#X_delivery_data = pd.concat([X_delivery_data,Onehot], axis=1)
```

```python
y_delivery_data = label_delivery_data['sectors']
```

```python
y_delivery_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(X_delivery_data,y_delivery_data, test_size=0.2,shuffle=True, random_state=42)
```

## 랜덤포레스트

```python
from sklearn.ensemble import RandomForestClassifier
delivery_forest = RandomForestClassifier(random_state=42, n_estimators=10, max_depth=50)
delivery_forest.fit(X_train, y_train)
```

### 1차 모델 평가

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred = delivery_forest.predict(X_train)
```

```python
accuracy_score(y_train, y_pred)
```

```python
# [테스트 샘플로 진행 시]
# accuracy_score
# max_depth = 50으로 진행했을 경우 (0.9874271887044267) 도출 : 교차검증을 진행하지 않았기 때문에 믿을 수 없는 값

# 단순히 레이블로 진행했을 경우 좋은 결과값 (0.5910842802439724) (max = 10)
# 원핫 인코딩을 진행했을 경우 오히려 안좋은 결과 값을 도출 (0.5215222082247141) (max=10)
```

```python
y_pred = delivery_forest.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
# 테스트 샘플로 진행 결과 0.64 정도로 나오게 됨
# 파라미터 조정 필요
```

### 하이퍼파라미터 조정

```python
from sklearn.model_selection import GridSearchCV
```

```python
from sklearn.ensemble import RandomForestClassifier
delivery_forest = RandomForestClassifier(random_state=42)
```

```python
forest_params ={'criterion':['gini','entropy']
                'max_depth':[10,20,30,40,50,60], 
                'n_estimators':[10,20,30]}
```

```python
gridserch_forest = GridSearchCV(delivery_forest, forest_params, scoring='accuracy', cv=5, n_jobs=-1)
```

```python
%time gridserch_forest.fit(X_train, y_train)
```

```python
gridserch_forest.best_params_
```

```python
# 결과별 스코어 확인
```

```python
cvres = gridserch_forest.cv_results_
```

```python
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(mean_score, params)
```

```python
# 1차 그리드 서치 결과
# {'criterion': 'entropy', 'max_depth': 20, 'n_estimators': 30}
# max_depth는 더 이상 늘려도 의미가 없지만 n_estimators는 좀더 큰 값을 줄 필요가 있음
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
delivery_forest = RandomForestClassifier(random_state=42)

forest_params ={'criterion':['gini','entropy'],
                'max_depth':[10,20,30,40,50,60], 
                'n_estimators':[30,40,50,60,70,80,90,100]}

gridserch_forest = GridSearchCV(delivery_forest, forest_params, scoring='accuracy', cv=5, n_jobs=-1)
```

```python
%time gridserch_forest.fit(X_train, y_train)
```

```python
gridserch_forest.best_params_
```

```python
cvres = gridserch_forest.cv_results_
```

```python
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(mean_score, params)
```

```python
## 하이퍼 파라미터 결과 0.670400261464967
# {'criterion': 'entropy', 'max_depth': 20, 'n_estimators': 90}
```
