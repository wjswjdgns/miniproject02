```python
delivery_data = pd.read_csv('datasets/delivery_data_mapplus.csv', encoding='utf-8')
```

```python
delivery_data['datetime'] = delivery_data['datetime'].apply(pd.to_datetime)
delivery_data['year'] = delivery_data['datetime'].apply(lambda x : x.year)
delivery_data['month'] = delivery_data['datetime'].apply(lambda x : x.month)
delivery_data['day'] = delivery_data['datetime'].apply(lambda x : x.day)
delivery_data = delivery_data.loc[:,['year','month','day','Day of the week','city','latitude','longitude','population','hour','rain','dust','humidity','precipitation','temp','windspeed','holiday','sectors']]
```

```python
label_delivery_data = delivery_data.copy()
label_length = len(label_delivery_data)
```

```python
for i in range(label_length):
    if label_delivery_data['Day of the week'][i] == 'Monday':
        label_delivery_data['Day of the week'][i] = 1
    elif label_delivery_data['Day of the week'][i] == 'Tuesday':
        label_delivery_data['Day of the week'][i] = 2
    elif label_delivery_data['Day of the week'][i] == 'Wednesday':
        label_delivery_data['Day of the week'][i] = 3
    elif label_delivery_data['Day of the week'][i] == 'Thursday':
        label_delivery_data['Day of the week'][i] = 4
    elif label_delivery_data['Day of the week'][i] == 'Friday':
        label_delivery_data['Day of the week'][i] = 5
    elif label_delivery_data['Day of the week'][i] == 'Saturday':
        label_delivery_data['Day of the week'][i] = 6
    elif label_delivery_data['Day of the week'][i] == 'Sunday':
        label_delivery_data['Day of the week'][i] = 7
```

```python
### 습도값 100인 행을 제거
label_delivery_data = label_delivery_data[label_delivery_data.humidity<100]
```

```python
### 바람 세기를 로그 변환
label_delivery_data['windspeed'] = np.log1p(label_delivery_data['windspeed'])

# 습도 log 변환 시 생기는 inf -inf 값을 nan값으로 변경
label_delivery_data['windspeed'] = label_delivery_data['windspeed'].replace([np.inf, -np.inf], np.nan)

# nan 값이 들어가 있는 행 삭제
label_delivery_data = label_delivery_data.dropna()
```

```python
columns_list = ['year','month','day','Day of the week','holiday', 'hour','latitude','longitude','population','rain', 'precipitation', 'humidity', 'temp', 'windspeed']
X_delivery_data = label_delivery_data[columns_list]
```

```python
y_delivery_data = label_delivery_data['sectors']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test, y_train, y_test = train_test_split(X_delivery_data,y_delivery_data, test_size=0.2,shuffle=True, stratify = y_delivery_data, random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestClassifier
delivery_forest = RandomForestClassifier(random_state=42, criterion='entropy')

forest_params ={ 'max_depth':[10,20,30,40,50,60],
                'n_estimators':[100,150,200],
                'min_samples_leaf':[2 ,4, 6, 8, 12, 18],
               'min_samples_split':[2 ,4, 6, 8, 16, 20]}

gridserch_forest = GridSearchCV(delivery_forest, forest_params, scoring='accuracy', cv=5, n_jobs=-1)
```

### 보깅 진행

```python
from sklearn.metrics import f1_score
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.ensemble import  RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

softmax_reg = LogisticRegression(multi_class='multinomial', C=0.01 ,random_state=42)
delivery_forest = RandomForestClassifier(random_state=42, criterion='entropy', max_depth=20, min_samples_leaf=2, min_samples_split=8, n_estimators=200)
log_reg = DecisionTreeClassifier(random_state=42, criterion='entropy', max_depth=10, min_samples_leaf=18, min_samples_split=2)
svm_clf = SVC(probability=True,random_state =42) #비선형구조로 probability 진행
```

```python
voting_clf = VotingClassifier(
    estimators = [('soft_rg', softmax_reg),('rf', delivery_forest),('tree', log_reg),('svm', svm_clf)],
    voting = 'soft'
)
```

```python
from sklearn.model_selection import cross_val_score
for clf in (softmax_reg, delivery_forest, log_reg, svm_clf,voting_clf):
    scores01 = cross_val_score(clf, X_train, y_train, scoring='accuracy', cv=5, n_jobs=-1)
    scores02 = cross_val_score(clf, X_train, y_train, scoring='f1_weighted', cv=5, n_jobs=-1)
    print(clf.__class__.__name__, "accuracy",scores01.mean(),"f1_weighted",scores02.mean())
```

```python
for clf in (softmax_reg, delivery_forest, log_reg, svm_clf,voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__,f1_score(y_test, y_pred, average='weighted'))
```
